```python
import random
def start_the_game():
  matrix = []
  for i in range(4):
      matrix.append([0]*4)
  return matrix

def adding_2(matrix):
  row = random.randint(0,3)
  column = random.randint(0,3)
  while(matrix[row][column] != 0):
      row = random.randint(0,3)
      column = random.randint(0,3)
  matrix[row][column] = 2
  return matrix

def reverse_matrix(matrix):
  matrix_2 = []
  for i in range(4):
      matrix_2.append([])
      for j in range(4):
          matrix_2[i].append(matrix[i][4-j-1])
  return matrix_2

def transpose_matrix(matrix): 
  matrix_2 = []
  for i in range(4):
      matrix_2.append([])
      for j in range(4):
          matrix_2[i].append(matrix[j][i])
  return matrix_2

def merge_matrix(matrix):
  changed_grid=False
  for i in range(4):
      for j in range(3):
          if matrix[i][j] == matrix[i][j+1] and matrix[i][j]!=0:
              matrix[i][j] = matrix[i][j]*2
              matrix[i][j+1] = 0
              changed_grid=True
  return matrix,changed_grid
            
def compress_matrix(matrix):
  changed_grid=False
  matrix_2 = []
  for i in range(4):
      matrix_2.append([0]*4)
  
  for i in range(4):
      pos=0
      for j in range(4):
          if matrix[i][j] != 0:
              matrix_2[i][pos] = matrix[i][j]
              if j!=pos:
                changed_grid=True
              pos+=1
  return matrix_2,changed_grid

def up(game_grid):
  transpose_game_grid = transpose_matrix(game_grid)
  new_game_grid,changed1 = compress_matrix(transpose_game_grid)
  new_game_grid,changed2 = merge_matrix(new_game_grid)
  changed_grid=changed1 or changed2
  new_game_grid = compress_matrix(new_game_grid)
  final_game_grid = transpose_matrix(new_game_grid)
  return final_game_grid,changed_grid

def down(game_grid):
  transpose_game__grid = transpose_matrix(game_grid)
  reversed_grid = reverse_matrix(transpose_game__grid)
  new_game_grid,changed1 = compress_matrix(reversed_grid)
  new_game_grid,changed2 = merge_matrix(new_game_grid)
  changed_grid=changed1 or changed2
  new_game_grid = compress_matrix(new_game_grid)
  final_reverse_game_grid = reverse_matrix(new_game_grid)
  final_game_grid = transpose_matrix(final_reverse_game_grid)
  return final_game_grid,changed_grid

def right(game_grid):
  reverse_game_grid = reverse_matrix(game_grid)
  new_game_grid,changed1 = compress_matrix(reverse_game_grid)
  new_game_grid,changed2 = merge_matrix(new_game_grid)
  changed_grid=changed1 or changed2
  new_game_grid = compress_matrix(new_game_grid)
  final_game_grid = reverse_matrix(new_game_grid)
  return final_game_grid,changed_grid

def left(game_grid):
  new_game_grid,changed1 = compress_matrix(game_grid)
  new_game_grid,changed2 = merge_matrix(new_game_grid)
  changed_grid=changed1 or changed2
  new_game_grid = compress_matrix(new_game_grid)
  return new_game_grid,changed_grid

    
def state_of_matrix(matrix):
  for i in range(4):
      for j in range(4):
          if (matrix[i][j] == 2048):
              return 'YOU WON'
  for i in range(4):
      for j in range(4):
          if (matrix[i][j] == 0):
              return 'PROCEED'
  for i in range(3):
      for j in range(3):
          if (matrix[i][j] == matrix[i+1][j] or matrix[i][j] == matrix[i][j+1]):
              return 'PROCEED'
  for j in range(3):
      if matrix[3][j] == matrix[3][j+1]:
          return 'PROCEED'
  for i in range(3):
      if matrix[i][3] == matrix[i+1][3]:
          return 'PROCEED'
      
  return 'YOU LOST'
```
